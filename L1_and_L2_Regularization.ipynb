# ***house price prediction***

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/Melbourne_housing_FULL.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop('Date',axis='columns')
```

```python
cols_to_use=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
df=df[cols_to_use]
df.head()
```

```python
df.isna().sum()
```

```python
colos_to_fill_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']
df[colos_to_fill_zero]=df[colos_to_fill_zero].fillna(0)
```

```python
df
```

```python
df.isna().sum()
```

```python
df['BuildingArea']=df.BuildingArea.fillna(df.BuildingArea.mean())
```

```python
df['Landsize']=df.Landsize.fillna(df.Landsize.mean())
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf=LinearRegression()
lr_clf.fit(x_train,y_train)
```

```python
lr_clf.score(x_train,y_train)
```

# ***L1 regularization***

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

# ***L2 regularization***

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```
