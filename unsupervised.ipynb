# ***KMean***

```python
import pandas as pd

data = {
    'Name': ['Rob', 'Michael', 'Mohan', 'Ismail', 'Kory', 'Gautam', 'David', 'Andrea',
             'Brad', 'Angelina', 'Donald', 'Tom', 'Arnold', 'Jared', 'Stark', 'Ranbir',
             'Dipika', 'Priyanka', 'Nick', 'Alia', 'Sid', 'Abdul'],

    'Age': [27, 29, 29, 28, 42, 39, 41, 38,
            36, 35, 37, 26, 27, 28, 29, 32,
            40, 41, 43, 39, 41, 39],

    'Income($)': [70000, 90000, 61000, 60000, 150000, 155000, 160000, 162000,
                  156000, 130000, 137000, 45000, 48000, 51000, 49500, 53000,
                  65000, 63000, 64000, 80000, 82000, 58000]
}

df = pd.DataFrame(data)
print(df)

```

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km=KMeans(n_clusters=3)
km
```

```python
km.fit_predict(df[['Age','Income($)']])
```

```python
y_pred=km.fit_predict(df[['Age','Income($)']])
y_pred
```

```python
df['cluster']=y_pred
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(df3.Age,df3['Income($)'],color='black')
plt.xlabel('Age')

plt.ylabel('Income($)')
plt.legend()
```

```python
scaler=MinMaxScaler()
scaler.fit_transform(df[['Income($)']])
df['Income($)']=scaler.fit_transform(df[['Income($)']])
df

```

```python
scaler.fit_transform(df[['Age']])
df['Age']=scaler.fit_transform(df[['Age']])
df
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['Age','Income($)']])
df['cluster']=y_predicted
df
```

```python
k_rng=range(1,11)
sse=[]
for k in k_rng:
 km=KMeans(n_clusters=k)
 km.fit(df[['Age','Income($)']])
 sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng,sse)
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df['target']=iris.target
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
```

```python
kmeans.fit(df.iloc[:, 0:4])
```

```python
df['cluster'] = kmeans.labels_
df.head()
```

```python
df
```

```python
comparison = df[['target', 'cluster']]
comparison.head(20)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['cluster'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Means Clustering (Iris)')
plt.show()
```

```python
mapping = {0:'setosa', 1:'versicolor', 2:'virginica'}
df['predicted_flower'] = df['cluster'].map(mapping)
df.head()
```

```python
pd.crosstab(df['target'], df['cluster'])
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Fit KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df.iloc[:, 0:4])

# Add cluster labels to DF
df['cluster'] = kmeans.labels_

# Plot clusters
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['cluster'])

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Means Clustering (Iris)')

# CENTROIDS
centers = kmeans.cluster_centers_

# plot centroids as red Xs
plt.scatter(
    centers[:, 0],        # centroid sepal length
    centers[:, 1],        # centroid sepal width
    c='red',
    s=300,
    marker='X'
)

plt.show()
```

```python
wcss = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df.iloc[:, 0:4])
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss, marker='o')
plt.xlabel('K')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

# **Naive bayes theorem**

```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
target=df.Survived
inputs=df.drop('Survived',axis='columns')
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
```

```python
df.head()
```

```python
inputs.drop('Sex',axis='columns',inplace=True)
inputs.head()
```

```python
inputs.Age[:10]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

# ***spam detection***

```python
import pandas as pd
df=pd.read_csv("spam.csv")
df.head()
```

```python
df.groupby("Category").describe()
```

```python
df['spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df.Message, df.spam, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
X_train_count=v.fit_transform(X_train)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count,Y_train)
X_test_count = v.transform(X_test)
```

# *** sklearn pipeline***

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```
