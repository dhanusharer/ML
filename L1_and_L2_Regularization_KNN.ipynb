# ***house price prediction***

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/Melbourne_housing_FULL.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop('Date',axis='columns')
```

```python
cols_to_use=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
df=df[cols_to_use]
df.head()
```

```python
df.isna().sum()
```

```python
colos_to_fill_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']
df[colos_to_fill_zero]=df[colos_to_fill_zero].fillna(0)
```

```python
df
```

```python
df.isna().sum()
```

```python
df['BuildingArea']=df.BuildingArea.fillna(df.BuildingArea.mean())
```

```python
df['Landsize']=df.Landsize.fillna(df.Landsize.mean())
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf=LinearRegression()
lr_clf.fit(x_train,y_train)
```

```python
lr_clf.score(x_train,y_train)
```

# ***L1 regularization***

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=100,max_iter=100,tol=0.1)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

# ***L2 regularization***

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=10,max_iter=10,tol=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
from sklearn.linear_model import LassoCV

model = LassoCV(alphas=[0.0001,0.001,0.01,0.1,1,10,100], cv=5)
model.fit(x_train, y_train)

print(model.alpha_)
```

# ***KNN***

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
df['target']=iris.target
```

```python
df.head()
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
x=df.drop('target',axis='columns')
y=df.target
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,Y_train)
knn.score(X_test,Y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(X_test)
cm=confusion_matrix(Y_test,y_pred)
cm
```

# ***digits prediction using KNN***

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
df=pd.DataFrame(digits.data,columns=digits.feature_names)
```

```python
df
```

```python
X=df
Y=digits.target
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
knn.fit(X_train,Y_train)
knn.score(X_test,Y_test)
```

```python
y_pred=knn.predict(X_test)
cm=confusion_matrix(Y_test,y_pred)
cm
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Build pipeline to scale + train KNN
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])

# Try different k values
param_grid = {
    'knn__n_neighbors': range(1, 31)   # test k = 1 to 30
}

grid = GridSearchCV(pipe, param_grid, cv=5)
grid.fit(X, Y)

print("Best k value:", grid.best_params_)
print("Best CV score:", grid.best_score_)
```
